```python
data1 = pd.read_csv('tips.csv')
data1
```

```python
# counting how many people have come to restrauent
sns.countplot(x='sex',data=data1)
plt.show()
```

```python
# using barplot
sns.barplot(x='sex', y='size', data=data1)
plt.show()
```

```python
# checking for smokers
# using hue='smoker'

sns.barplot(x='sex', y='size', data=data1, hue='smoker')
plt.show()
```

```python
# converting this to horizontally by replacing the values
sns.countplot(y='sex', data=data1, hue='smoker')
plt.show()
```

```python
# changing colors using
# palette='value'

sns.countplot(x='sex', data=data1, hue='smoker', palette='rainbow')
plt.show()
```

```python
# changing colors using
# color='value'

sns.countplot(x='sex', data=data1, hue='smoker', color='g')
plt.show()
```

```python
# giving above diagram a dark or light color using 
# saturation = float (0 to 1)

sns.countplot(x='sex', data=data1, hue='smoker', color='g',
             saturation=0.1)
plt.show()
```
