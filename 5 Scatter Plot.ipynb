```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from IPython import display
```

# Scatter Plot

```python
display.Image('img4.png')
```

```python
# getting data from github
data1 = sns.load_dataset('penguins')
data1
```

```python
# plotting scatter plot

sns.scatterplot(x=data1.bill_length_mm, y=data1.bill_depth_mm)
plt.show()
```

```python
# differentiating data by sex
# using hue='sex'

sns.scatterplot(x=data1.bill_length_mm, y=data1.bill_depth_mm,
               hue=data1.sex)
plt.show()
```

```python
# the above graph is using big data set
# we can small it by using given example

data2 = sns.load_dataset('penguins').head(20)
data2
```

```python
# now plotting scatter plot for small data
sns.scatterplot(x=data2.bill_length_mm, y=data2.bill_depth_mm,
               hue=data2.sex)
plt.show()
```

```python
# in above graph male and female have dots
# we can change them for both using
# style='sex'

sns.scatterplot(x=data2.bill_length_mm, y=data2.bill_depth_mm,
               hue=data2.sex, style=data2.sex)
plt.show()
```

```python
# we can change dot sizes also
# using size='sex'

sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=data2,
               hue='sex', style='sex', size='sex')
plt.show()
```

```python
# adjusting size according to our need
# using sizes = (int,int)

sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=data2,
               hue='sex', style='sex', size='sex', sizes=[60,100])
plt.show()
```

```python
# changing colors
# using paletter = 'value'

sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=data2,
               hue='sex', style='sex', size='sex', sizes=[60,100],
               palette='terrain')
plt.show()
```

```python
# we can give light brightness these colors
# using alpha=float

sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=data2,
               hue='sex', style='sex', size='sex', sizes=[60,100],
               palette='terrain', alpha=0.6)
plt.show()
```

```python
# we can change the markers also
# using markers = {'val1':'shape1', 'val2':'shape2'}

m={'Male':'o', "Female":'*'}

sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=data2,
               hue='sex', style='sex', size='sex', sizes=[60,100],
               palette='terrain', markers=m)
plt.show()
```

```python
# using large data

m={'Male':'o', "Female":'*'}

sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=data1,
               hue='sex', style='sex', size='sex', sizes=[60,100],
               palette='terrain', markers=m)
plt.show()
```
