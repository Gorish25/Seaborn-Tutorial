```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from IPython import display
```

# Strip Plot

```python
display.Image('img10.png')
```

```python
data1 = pd.read_csv('tips.csv')
data1
```

```python
# creating strip plot
sns.stripplot(x='day',y='total_bill',data=data1)
plt.show()
```

```python
# differentiating using
# hue='sex'

sns.stripplot(x='day',y='total_bill',data=data1,hue='sex')
plt.show()
```

```python
# changing colors using
# paletter='value'

sns.stripplot(x='day',y='total_bill',data=data1,hue='sex',
             palette='rocket_r')
plt.show()
```

```python
# changing the linewidth
# using linewidth=integer

sns.stripplot(x='day',y='total_bill',data=data1,hue='sex',
             palette='rocket_r',linewidth=2)
plt.show()
```

```python
# changing the edge color
# using edgecolor='value'

sns.stripplot(x='day',y='total_bill',data=data1,hue='sex',
             palette='rocket_r',linewidth=2, edgecolor='g')
plt.show()
```

```python
# distributing values in all graph using
# jitter=integer

sns.stripplot(x='day',y='total_bill',data=data1,hue='sex',
             palette='rocket_r',linewidth=2, edgecolor='g',
             jitter=5)
plt.show()
```

```python
# increasing the sizes of dots using
# size=integer

sns.stripplot(x='day',y='total_bill',data=data1,hue='sex',
             palette='rocket_r',linewidth=2, edgecolor='g',
             size=10)
plt.show()
```

```python
# changing markers using
# marker='type'

m={'Male':'o', 'Female':'<'}
sns.stripplot(x='day',y='total_bill',data=data1,hue='sex',
             palette='rocket_r',linewidth=2, edgecolor='g',
             size=10,marker='<')
plt.show()
```

```python
# giving a light look using
# alpha=float (0 to 1)

sns.stripplot(x='day',y='total_bill',data=data1,hue='sex',
             palette='rocket_r',linewidth=2, edgecolor='g',
             size=10,marker='<', alpha=0.5)
plt.show()
```

```python
# strip plot of only a single column

sns.stripplot(x='total_bill',data=data1)
plt.show()
```

```python
# changing above graph to vertically

sns.stripplot(y='total_bill',data=data1)
plt.show()
```
