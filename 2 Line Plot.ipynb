```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# creating data
var1 = [1,2,3,4,5,6,7]
var2 = [2,3,4,1,5,6,7]
```

```python
# creating a matplotlib graph
plt.plot(var1,var2)
plt.show()
```

```python
# creating a seaborn graph
# 1. use lintplot
# 2. we need a dataframe also so importing pandas and creating
#    and dataframe

import pandas as pd

# creating values for lineplot
var1 = [1,2,3,4,5,6,7]
var2 = [2,3,4,1,5,6,7]

# creating a dataframe
x = pd.DataFrame({'var1':var1, 'var2':var2})

sns.lineplot(x=var1,y=var2, data=x)
```

```python
# in upper graph a label is coming above the graph 
# which is <AxesSubplot:xlabel='var1', ylabel='var2'>
# we can remove this using plt.show()

# creating values for lineplot
var1 = [1,2,3,4,5,6,7]
var2 = [2,3,4,1,5,6,7]

# creating a dataframe
x = pd.DataFrame({'var1':var1, 'var2':var2})

sns.lineplot(x=var1,y=var2, data=x)

# removing the label
plt.show()

# in given graph label on x-axis (var1) and y-axis (var2)
# are automatically coming
```

```python
# loading some external data using
# sns.load_dataset()

# loading external data from seaborn official github
x1 = sns.load_dataset('penguins')
x1
```

```python
# plotting linear plot by using above dataset

# loading dataset
x1 = sns.load_dataset('penguins')

# plotting it to linear plot (a specific column)
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=x1)
plt.show()
```

```python
# another example
sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=x1)
plt.show()
```

```python
# creating graph according to sex
# using hue='sex'
sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=x1,
            hue='sex')
plt.show()
```

```python
# changing style of lines using 
# style='sex'

sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=x1,
            hue='sex', style='sex')
plt.show()
```

```python
# changing colors using
# using palette='value'

sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=x1,
            hue='sex', style='sex',palette='Accent')
plt.show()
```

```python
# giving marker for showing the points using
# markers=['o','>'] 
# providing 2 values because graph has 2 genders male and female

sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=x1,
            hue='sex', style='sex',palette='Accent', 
            markers=['o','>'])
plt.show()
```

```python
# in the above dataset is very big
# lowering it using
# .head(size)

x1 = sns.load_dataset('penguins').head(20)
x1
```

```python
# now applying the lineplot
sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=x1,
            hue='sex', style='sex',palette='Accent', 
            markers=['o','>'])
plt.show()
```

```python
# in above graph has some dotted line
# we can remove it by using
# dashes = False
# by default is True

sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=x1,
            hue='sex', style='sex',palette='Accent', 
            markers=['o','>'], dashes=False)
plt.show()
```

```python
# in above graph legend is showing 
# we can remove it by using 
# legend=False
# by default is True

sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=x1,
            hue='sex', style='sex',palette='Accent', 
            markers=['o','>'], dashes=False, legend=False)
plt.show()
 
#  legend also have some values to change
```

```python
# giving above graph a grid

sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=x1,
            hue='sex', style='sex',palette='Accent', 
            markers=['o','>'], dashes=False, legend=False)

plt.grid()

plt.show()
 
```

```python
# giving graph a title

sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=x1,
            hue='sex', style='sex',palette='Accent', 
            markers=['o','>'], dashes=False, legend=False)

plt.grid()
plt.title('Linear Plot using Seaborn')

plt.show()
 
```
