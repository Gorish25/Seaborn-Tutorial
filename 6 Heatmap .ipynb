```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from IPython import display
```

```python
# creating a 2D Array using NumPy
var1 = np.linspace(1,10,20).reshape(4,5)
var1
```

```python
# converting above 2D array to heatmap
sns.heatmap(var1)
plt.show()
```

```python
# loading data from Github

data1 = sns.load_dataset('anagrams')
data1
```

```python
# we cant create string values heatmap (not allowed)

# 1. first we need to drop some data
data2 = data1.drop(columns=['attnr'], axis=1)
data2
```

```python
# creating heatmap of above table
sns.heatmap(data2)
plt.show()
```

```python
# above heatmap is for large data 
# creating heatmap for small data
data3 = data2.head(10)
sns.heatmap(data3)
plt.show()
```

```python
# changing minimum and maximum values in heatmap
sns.heatmap(data3, vmin=0, vmax=12)
plt.show()
```

```python
# changing side bar colors using 
# cmap='value'
sns.heatmap(data3, vmin=0, vmax=12, cmap='RdPu')
plt.show()
```

```python
# showing every colors position value
# using annot=True
sns.heatmap(var1, vmin=0, vmax=10, annot=True)
plt.show()
```

```python
# giving less values and creating a heatmap
var2 = np.linspace(1,10,10).reshape(2,5)
var2
```

```python
# creating heatmap of above values
sns.heatmap(var2, vmin=0,vmax=10,annot=True)
plt.show()
```

```python
# another example of 2D array
# these are the numbers for postion on above diagram
var3 = np.array([['a0','a1','a2','a3','a4'],
                ['b0','b1','b2','b3','b4']])
var3
```

```python
# creating heatmap of above array
# in this we have to change annot=var3
# we have to also specify that it is a string data 
# using fmt='s'
# otherwise error comes

sns.heatmap(var2, vmin=0, vmax=10, cmap='PuOr',
            annot=var3, fmt='s')
plt.show()
```

```python
# we can change the colors and font of given strings or any
# numbers coming on color boxes 

# giving values for font and color
y = {'fontsize':20, 'color':'r'}


# now we use annot_kws = y

sns.heatmap(var2, vmin=0, vmax=10, cmap='PuOr',
            annot=var3, fmt='s', annot_kws=y)
plt.show()
```

```python
# giving gap to each box using 
# linewidths=integer

sns.heatmap(var2, vmin=0, vmax=10, cmap='PuOr',
            annot=var3, fmt='s', annot_kws=y, linewidths=4)
plt.show()
```

```python
# we can give seperator a color also

sns.heatmap(var2, vmin=0, vmax=10, cmap='PuOr',
            annot=var3, fmt='s', annot_kws=y, linewidths=10,
           linecolor='g')
plt.show()
```

```python
# we can remove side color bar also using
# cbar=False

sns.heatmap(var2, vmin=0, vmax=10, cmap='PuOr',
            annot=var3, fmt='s', annot_kws=y, linewidths=10,
           linecolor='g', cbar=False)
plt.show()
```

```python
# we can remove x-axis and y-axis labels also 
# using xticklabels=False
# yticklabels=False

sns.heatmap(var2, vmin=0, vmax=10, cmap='PuOr',
            annot=var3, fmt='s', annot_kws=y, linewidths=10,
           linecolor='g', cbar=False, xticklabels=False,
           yticklabels=False)
plt.show()
```

```python
# we can set our own label also using
# data1.set(xlabel='value', ylabel='value2')

d = sns.heatmap(var2, vmin=0, vmax=10, cmap='PuOr',
            annot=var3, fmt='s', annot_kws=y, linewidths=10,
           linecolor='g', cbar=False, xticklabels=False,
           yticklabels=False)

d.set(xlabel='Python', ylabel='Learning')


plt.show()
```

```python
# we can increase or decrease labels font also
# using sns.set(font_scale=integer)

d = sns.heatmap(var2, vmin=0, vmax=10, cmap='PuOr',
            annot=var3, fmt='s', annot_kws=y, linewidths=10,
           linecolor='g', cbar=False, xticklabels=False,
           yticklabels=False)

d.set(xlabel='Python', ylabel='Learning')

sns.set(font_scale=5)

plt.show()
```
