```python
# creating dataframe
data1 = pd.read_csv('tips.csv')
data1
```

```python
# creating a catplot using tip and size

sns.catplot(x='tip',y='size',data=data1)
plt.show()
```

```python
# catplot for small data

data2 = pd.read_csv('tips.csv').head(20)

sns.catplot(x='tip',y='size',data=data2)
plt.show()
```

```python
# differentiating male and female using
# hue='sex'

sns.catplot(x='tip',y='size',data=data2, hue='sex')
plt.show()
```

```python
# changing colors using 
# palette='value'

sns.catplot(x='tip',y='size',data=data2, hue='sex', 
            palette='flare')
plt.show()
```

```python
# increasing and decreasing the height of graph
# using height=integer

sns.catplot(x='tip',y='size',data=data2, hue='sex', 
            palette='flare', height=3)
plt.show()
```

```python
# changing type of catplot using
# kind='type'

sns.catplot(x='tip',y='size',data=data2, hue='sex', 
            palette='flare', kind='bar')
plt.show()
```

```python
# kind='point'

sns.catplot(x='tip',y='size',data=data2, hue='sex', 
            palette='flare', kind='point')
plt.show()
```
