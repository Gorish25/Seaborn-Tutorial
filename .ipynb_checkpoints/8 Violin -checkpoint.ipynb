```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from IPython import display
```

# Violin

```python
display.Image('img8.png')
```

```python
# loading dataset 
data1 = pd.read_csv('tips.csv')
data1
```

```python
# creating violin plot
sns.violinplot(x='day', y='total_bill', data=data1)
plt.show()
```

```python
# differentiating using 
# hue='time'

sns.violinplot(x='day', y='total_bill', data=data1,
              hue='time')
plt.show()
```

```python
# increasing or decreasing linewidth

sns.violinplot(x='day', y='total_bill', data=data1,
              hue='time', linewidth=5)
plt.show()
```

```python
# giving different colors using 
# paletter = 'value'

sns.violinplot(x='day', y='total_bill', data=data1,
              hue='time', palette='Set2')
plt.show()
```

```python
# differentiating data according to Lunch and Dinner
# using order=['Lunch','Dinner']

sns.violinplot(x='time', y='total_bill', data=data1,
            order=['Lunch','Dinner'])
plt.show()
```

```python
# giving graph a light or dark look using 
# saturation=float (0 to 1)

sns.violinplot(x='day', y='total_bill', data=data1,
              saturation=0.3)
plt.show()
```

```python
# changing all graph color
# using color='value'

sns.violinplot(x='day', y='total_bill', data=data1,
              color='r')
plt.show()
```

```python
# splitting graph to male and female
# using hue='sex', split=True

sns.violinplot(x='day', y='total_bill', data=data1,
              hue='sex', split=True)
plt.show()
```

```python
# changes making according to scale

sns.violinplot(x='day', y='total_bill', data=data1,
              hue='sex', split=True, scale='width')
plt.show()
```

```python
# changing above graph to horizontaly

sns.violinplot(y='day', x='total_bill', data=data1,
              hue='sex', split=True, scale='width')
plt.show()
```

```python
# changing the violin type
# using inner='value'
# values of inner = ['box','quartile','point','stick',None]
# by default it 'box'

sns.violinplot(x='time', y='total_bill', data=data1,
              order=['Lunch','Dinner'], inner='quart')
```

```python
# creating a single violin graph

sns.violinplot(x=data1.total_bill)
plt.show()
```

```python
# along with y axis

sns.violinplot(y=data1.total_bill)
plt.show()
```
