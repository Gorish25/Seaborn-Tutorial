```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from IPython import display
```

# Pair Plot

```python
display.Image('img9.png')
```

```python
data1 = pd.read_csv('tips.csv')
data1
```

```python
# creating a pairplot

sns.pairplot(data1)
plt.show()
```

```python
# differentiating according to sex
# using hue='sex'

sns.pairplot(data1,hue='sex')
plt.show()
```

```python
# seeing only specific data
# using vars=['val1','val2']

sns.pairplot(data1,hue='sex',vars=['total_bill','tip'])
plt.show()
```

```python
# changing the hue order (female first)
# using hue_order=['Female','Male']

sns.pairplot(data1,hue='sex',vars=['total_bill','tip'],
            hue_order=['Female','Male'])
plt.show()
```

```python
# changing colors using
# paletter = 'value'

sns.pairplot(data1,hue='sex',vars=['total_bill','tip'],
            hue_order=['Female','Male'], palette='afmhot')
plt.show()
```

```python
# giving total_bill and tip to x-axis
# using x_vars=['total_bill','tip']

sns.pairplot(data1,hue='sex',x_vars=['total_bill','tip'],
            hue_order=['Female','Male'], palette='afmhot')
plt.show()
```

```python
# changing the kind of graph showing
# using kind='value'
# values = ['scatter','kde','hist','reg']

sns.pairplot(data1, hue='sex', hue_order=['Female','Male'],
            kind='reg')

plt.show()
```

```python
# another example 

sns.pairplot(data1, hue='sex', hue_order=['Female','Male'],
            kind='kde')

plt.show()
```

```python
# giving diagonals histogram
# using 

sns.pairplot(data1, hue='sex', hue_order=['Female','Male'],
            kind='kde', diag_kind='hist')

plt.show()
```

```python
# changing markers using
# markers=['type1','type2']

sns.pairplot(data1, hue='sex', hue_order=['Female','Male'],
            markers=['*','<'])

plt.show()
```
