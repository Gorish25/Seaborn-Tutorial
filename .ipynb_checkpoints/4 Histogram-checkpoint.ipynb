```python
import matplotlib.pyplot as plt
import seaborn as sns
from IPython import display
import pandas as pd
import numpy as np
```

# Histogram

```python
display.Image('img3.png')
```

```python
# getting data from github

data1 = sns.load_dataset('penguins')
data1
```

```python
# plotting histogram
# requires 1 value
sns.displot(data1['bill_length_mm'])
plt.show()
```

```python
# another histogram
sns.displot(data1['flipper_length_mm'])
plt.show()
```

```python
# in above graph x axis values are not orgainised well 
# we can orgainise them using
# bins = [val1,val2, .... , valn]

sns.displot(data1['flipper_length_mm'], bins=[170,180,190,200,210,220,230])
plt.show()
```

```python
# we can use a line above these graphs using
# kde = True (kernal density estimate)

sns.displot(data1['flipper_length_mm'], 
            bins=[170,180,190,200,210,220,230], kde=True)
plt.show()
```

```python
# we can also give some scale lines to x-axis using 
# rug = True

sns.displot(data1['flipper_length_mm'],
            bins=[170,180,190,200,210,220,230], kde=True,
           rug=True)
plt.show()
```

```python
# changing the colors of bars using
# color = 'value'

sns.displot(data1['flipper_length_mm'],
            bins=[170,180,190,200,210,220,230], kde=True,
           rug=True, color='g')
plt.show()
```

```python
# we can also show data using log scale
# log_scale = True

sns.displot(data1['flipper_length_mm'], kde=True, log_scale=True,
           rug=True, color='g')
plt.show()
```
