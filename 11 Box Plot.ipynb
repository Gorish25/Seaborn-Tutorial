```python
data1 = pd.read_csv('tips.csv')
data1
```

```python
# creating a grid for box plot
sns.set(style='whitegrid')

# plotting box plot accroding to days
sns.boxplot(x='day',y='total_bill',data=data1)
plt.show()
```

```python
# differentiating according to time
# using hue='time'

sns.boxplot(x='day',y='total_bill',data=data1, hue='time')
plt.show()
```

```python
# differentiating according to sex
# using hue='sex'

sns.boxplot(x='day',y='total_bill',data=data1,hue='sex')
plt.show()
```

```python
# changing color
# using color='value'

sns.boxplot(x='day',y='total_bill',data=data1, hue='sex',
           color='g')
plt.show()
```

```python
# giving order 
# using order=['val1','val2']

sns.boxplot(x='day',y='total_bill',data=data1, hue='sex',
           color='g',order=['Thur','Fri','Sat','Sun'])
plt.show()
```

```python
# showing means in every box
# using showmeans=True

sns.boxplot(x='day',y='total_bill',data=data1, hue='sex',
           color='r',order=['Thur','Fri','Sat','Sun'],
           showmeans=True)
plt.show()
```

```python
# mean properties
# using meanprops={}

sns.boxplot(x='day',y='total_bill',data=data1, hue='sex',
           color='g',order=['Thur','Fri','Sat','Sun'],
           showmeans=True, 
            meanprops={'marker':'o','markeredgecolor':'r'})
plt.show()
```

```python
# increasing linewidth of box
# using linewidth=integer

sns.boxplot(x='day',y='total_bill',data=data1, hue='sex',
           color='g',order=['Thur','Fri','Sat','Sun'],
           showmeans=True, 
            meanprops={'marker':'o','markeredgecolor':'r'},
           linewidth=5)
plt.show()
```

```python
# chaning colors using 
# palette='value'

sns.boxplot(x='day',y='total_bill',data=data1, hue='sex',
           palette='cividis',order=['Thur','Fri','Sat','Sun'],
           showmeans=True, 
            meanprops={'marker':'o','markeredgecolor':'r'},
           linewidth=5)
plt.show()
```

```python
# changing its orientation using
# orient='h' or 'v'
# if we want to convert it into horizontal then
# y axis must have numeric data
# we can change axis also to cover this problem

sns.boxplot(y='day',x='total_bill',data=data1, hue='sex',
           order=['Thur','Fri','Sat','Sun'],
           showmeans=True, 
            meanprops={'marker':'o','markeredgecolor':'r'},
           linewidth=5)
plt.show()
```

```python
# using orient by not specifying x and y axis

sns.boxplot(data=data1,
           showmeans=True, 
            meanprops={'marker':'o','markeredgecolor':'r'},
           linewidth=5,orient='h')
plt.show()
```

```python
# box plot for a single column

sns.boxplot(x=data1.total_bill)
plt.show()
```

```python
# changing it to vertical
# by changing to y

sns.boxplot(y=data1.total_bill)
plt.show()
```
