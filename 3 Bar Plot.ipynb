```python
import matplotlib.pyplot as plt
import seaborn as sns
from IPython import display
import pandas as pd
import numpy as np
```

# Bar Plot

```python
# getting large amount of data from github using  
# seaborn datasheet

data1 = sns.load_dataset('penguins')
data1
```

```python
# plotting a bar plot using above datasheet

sns.barplot(x=data1.island, y=data1.bill_length_mm)
```

```python
# removing the above label (<AxesSubplot:xlabel='island', ylabel='bill_length_mm'>)
# using plt.show()

sns.barplot(x=data1.island, y=data1.bill_length_mm)
plt.show()
```

```python
# if we not use data1.island
# then another method is 

sns.barplot(x='island', y='bill_length_mm', data=data1)
plt.show()
```

```python
# creating a double bar graph using
# hue='sex' ('' => have any specific value)
# hue = data_set.sex

sns.barplot(x=data1.island, y=data1.bill_length_mm, hue=data1.sex)
plt.show()
```

```python
# changing the order of x axis values

order2 = ["Dream",'Torgersen','Biscoe']
sns.barplot(x=data1.island, y=data1.bill_length_mm, hue=data1.sex,
           order=order2)
plt.show()
```

```python
# changing male and female graph position
# using hue_order=['Female','Male']

order2 = ["Dream",'Torgersen','Biscoe']
sns.barplot(x=data1.island, y=data1.bill_length_mm, hue=data1.sex,
           order=order2, hue_order=['Female','Male'])
plt.show()
```

```python
# above bar graphs having a line on top we can remove it
# or increase or decrease it using
# ci = integer (must be between 0 to 100)

order2 = ["Dream",'Torgersen','Biscoe']
sns.barplot(x=data1.island, y=data1.bill_length_mm, hue=data1.sex,
           order=order2, hue_order=['Female','Male'], ci=0)
plt.show()
```

```python
# using n_boot = integer

order2 = ["Dream",'Torgersen','Biscoe']
sns.barplot(x=data1.island, y=data1.bill_length_mm, hue=data1.sex,
           order=order2, hue_order=['Female','Male'], n_boot=5)
plt.show()
```

```python
# we can change its orientation also 
# using orient =  "h" or 'v'
# but use when x and y axis both have float or int value


# horizontal orientation
sns.barplot(x=data1.bill_length_mm, y=data1.bill_depth_mm, 
           orient='h')
plt.show()
```

```python
# vertical
sns.barplot(x=data1.bill_length_mm, y=data1.bill_depth_mm, 
           orient='v')
plt.show()
```

```python
# changing colors of bars

sns.barplot(x=data1.island, y=data1.bill_length_mm, hue=data1.sex,
           order=order2, hue_order=['Female','Male'], 
           color='g')
plt.show()
```

```python
# list of colors are in 
# palatte = 'value'

sns.barplot(x=data1.island, y=data1.bill_length_mm, hue=data1.sex,
           order=order2, hue_order=['Female','Male'], 
           color='g', palette='brg_r')
plt.show()
```

```python
# above graph is a bit dark
# we can give light colors using 
# saturation = float

sns.barplot(x=data1.island, y=data1.bill_length_mm, hue=data1.sex,
           order=order2, hue_order=['Female','Male'], 
           color='g', palette='brg_r', saturation=0.9)
plt.show()
```

```python
# on the top of bars a line is showing
# we can change its color also using
# errcolor = 'value'

sns.barplot(x=data1.island, y=data1.bill_length_mm, hue=data1.sex,
           order=order2, hue_order=['Female','Male'], 
           color='g', palette='brg_r', saturation=0.9,
           errcolor='g')
plt.show()
```

```python
# we can increase or decrease its width also using
# errwidth = integer

sns.barplot(x=data1.island, y=data1.bill_length_mm, hue=data1.sex,
           order=order2, hue_order=['Female','Male'], 
           color='g', palette='brg_r', saturation=0.9,
           errcolor='g', errwidth=15)
plt.show()
```

```python
# we can give bars a cap also 
# using capsize=float

sns.barplot(x=data1.island, y=data1.bill_length_mm, hue=data1.sex,
           order=order2, hue_order=['Female','Male'], 
           color='g', palette='brg_r', saturation=0.9,
           errcolor='g', capsize=0.6)
plt.show()
```

```python
# both bars (male female) can be one upon another using
# dodge = False
# by default is True

sns.barplot(x=data1.island, y=data1.bill_length_mm, hue=data1.sex,
           order=order2, hue_order=['Female','Male'], 
           color='g', palette='brg_r', saturation=0.9,
           errcolor='g', dodge=False)
plt.show()
```

```python
# we can decrese its brightness using 
# alpha = float

sns.barplot(x=data1.island, y=data1.bill_length_mm, hue=data1.sex,
           order=order2, hue_order=['Female','Male'], 
           color='g', palette='brg_r', saturation=0.9,
           errcolor='g', alpha=0.4)
plt.show()
```

```python
# giving a grid using
# plt.grid()

sns.barplot(x=data1.island, y=data1.bill_length_mm, hue=data1.sex,
           order=order2, hue_order=['Female','Male'], 
           color='g', palette='brg_r', saturation=0.9,
           errcolor='g')

plt.grid()
plt.show()
```

```python
# seaborn also have its own grid 
# we can use it by using
# sns.set(style='value')
# it must be on top

sns.set(style='darkgrid')

sns.barplot(x=data1.island, y=data1.bill_length_mm, hue=data1.sex,
           order=order2, hue_order=['Female','Male'], 
           color='g', palette='brg_r', saturation=0.9,
           errcolor='g')
plt.show()
```
