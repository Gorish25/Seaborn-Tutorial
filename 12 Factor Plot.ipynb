```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from IPython import display
```

# Factor Plot

```python
display.Image('img13.png')
```

```python
data1 = pd.read_csv('tips.csv')
data1
```

```python
# plotting graph between tip and size
sns.factorplot(x='size',y='tip',data=data1)
plt.show()
```

```python
# differentiating according to sex

sns.factorplot(x='size',y='tip',data=data1, hue='sex')
plt.show()
```

```python
# we can create multiple graphs also using
# kind='type of graph'

# example kind='bar'
sns.factorplot(x='size',y='tip',data=data1, hue='sex', kind='bar')
plt.show()
```

```python
# kind='box' 

sns.factorplot(x='size',y='tip',data=data1, hue='sex', kind='box')
plt.show()
```
