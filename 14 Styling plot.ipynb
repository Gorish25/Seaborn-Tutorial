```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from IPython import display
```

```python
data1 = pd.read_csv('tips.csv')
data1
```

```python
# box plot of total_bill (vertical)
sns.boxplot(y='total_bill', data=data1)
plt.show()
```

```python
# bar plot of total_bill in comparison of day
sns.barplot(x='day',y='total_bill', data=data1)
plt.show()
```

```python
# using grid
sns.barplot(x='day',y='total_bill', data=data1)
plt.grid()
plt.show()
```

```python
# changing style
# using sns.set_style()

sns.set_style('dark')
sns.barplot(x='day',y='total_bill', data=data1)
plt.show()
```

```python
# darkgrid

sns.set_style('darkgrid')
sns.barplot(x='day',y='total_bill', data=data1)
plt.show()
```

```python
# removing top and right line using 
# sns.dispine()

sns.set_style('white')
sns.barplot(x='day',y='total_bill', data=data1)

sns.despine()

plt.show()
```

```python
# increasing size using
# plt.figure(figsize=(width,height))

sns.set_style('white')

plt.figure(figsize=(12,7))

sns.barplot(x='day', y='total_bill', data=data1)

sns.despine()
plt.show()
```

```python
# another example

sns.set_style('white')

plt.figure(figsize=(3,12))

sns.barplot(x='day', y='total_bill', data=data1)

sns.despine()
plt.show()
```

```python
# setting context
# using sns.set_context('type',font_scale=int,float)
# type => poster, paper

sns.set_style('white')

sns.set_context('poster',font_scale=0.8)

sns.barplot(x='day', y='total_bill', data=data1)

plt.show()
```

```python
# changing colors
# using palette = 'value'

sns.set_style('white')

sns.set_context('poster',font_scale=0.8)

sns.barplot(x='day', y='total_bill', data=data1, palette='gist_yarg')

plt.show()
```
